# 開始執行處理

```
# _*_ coding: utf-8 _*_
# _*_ coding: cp950 _*_

# author: Chi-Hsu Chen (css920@gmail.com)
# datetime: 20200602

import numpy as np
import cv2
import time
from google.colab.patches import cv2_imshow
```

```
# 載入圖片
imgLena=cv2.imread('sample_data/lena.png',cv2.IMREAD_COLOR)
imgLenaGray=cv2.cvtColor(imgLena,cv2.COLOR_BGR2GRAY)

cv2_imshow(imgLenaGray)

# 建立SIFT偵測器
detectorSIFT=cv2.xfeatures2d_SIFT.create()
keypoints=detectorSIFT.detect(imgLenaGray,None)

# 使用drawKeypoints劃出關鍵點
imgKeypoints=cv2.drawKeypoints(imgLenaGray,keypoints,imgLena)

cv2_imshow(imgKeypoints)
```
