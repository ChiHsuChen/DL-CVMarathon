```
# 顯示原圖
imgLenaColor=cv2.imread('sample_data/lena.png')
cv2_imshow(imgLenaColor)
```

# 透視轉換

```
img_perspective = imgLenaColor.copy()
height, width = imgLenaColor.shape[:2]

# 設定四對點，並取得 perspective 矩陣
point1 = np.array([[60, 40], [420, 40], [420, 510], [60, 510]], dtype=np.float32)
point2 = np.array([[0, 80], [width, 120], [width, 430], [0, 470]], dtype=np.float32)
M=cv2.getPerspectiveTransform(point1, point2)

imgWarpPerspective=cv2.warpPerspective(imgLenaColor,M,(height,width))

cv2_imshow(imgWarpPerspective)
```
